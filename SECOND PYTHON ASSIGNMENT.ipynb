```python
print("===== SIMPLE CALCULATOR =====")
first_number = float(input("Enter first number: "))
calc_operator = input("Enter operator (+, -, *, /, //, %, **): ")
second_number = float(input("Enter second number: "))

calculation_error = False

if second_number == 0 and calc_operator in ["/", "//", "%"]:
    calculation_error = True
    print("Error: You cannot divide by zero.")
elif calc_operator == "+":
    total = first_number + second_number
elif calc_operator == "-":
    total = first_number - second_number
elif calc_operator == "*":
    total = first_number * second_number
elif calc_operator == "/":
    total = first_number / second_number
elif calc_operator == "//":
    total = int(first_number // second_number)
elif calc_operator == "%":
    total = int(first_number % second_number)
elif calc_operator == "**":
    total = first_number ** second_number
else:
    calculation_error = True
    print("Error: That is an invalid operator.")

if calculation_error == False:
    if calc_operator == "/":
        print(f"\nResult: {first_number} / {second_number} = {total:.4f}")
    else:
        print(f"\nResult: {first_number} {calc_operator} {second_number} = {total}")

    if total > 0:
        print("The result is positive.")
    elif total < 0:
        print("The result is negative.")
    else:
        print("The result is zero.")

print("==============================")
```
