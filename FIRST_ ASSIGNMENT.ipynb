```python
print("===== Enter Student Details =====")
name_input = input("Full name: ")
age_input = input("Age: ")
gpa_input = input("GPA: ")
courses_input = input("Courses enrolled (separate with commas): ")

student_name = name_input
student_age = int(age_input)
student_gpa = float(gpa_input)

student_courses = [course.strip() for course in courses_input.split(",")]

is_on_deans_list = student_gpa >= 3.5
is_a_minor = student_age < 18

deans_list_status = "Yes" if is_on_deans_list else "No"
age_status = "Minor" if is_a_minor else "Adult"

student_profile = {
    "name": student_name,
    "age": student_age,
    "gpa": student_gpa,
    "courses": student_courses,
    "num_courses": len(student_courses),
}

print("\n===== STUDENT PROFILE =====")
print(f"Name:         {student_profile['name']}")
print(f"Age:          {student_profile['age']} ({age_status})")
print(f"GPA:          {student_profile['gpa']}")
print(f"Courses:      {student_profile['courses']}")
print(f"# Courses:    {student_profile['num_courses']}")
print(f"Dean's List:  {deans_list_status}")
print("===========================")
```
